```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

# 替换这里的路径为你自己的图片路径
image_path = "misc/keySen/encrypt_image_peper256.png"

# 打开图像并转换为灰度图像
img = Image.open(image_path).convert('L')

# 将图像数据转换为numpy数组
img_array = np.array(img)

# 计算直方图
histogram, bin_edges = np.histogram(img_array, bins=256, range=(0, 256))

# 绘制直方图
plt.figure()
plt.title("Grayscale Histogram")
plt.xlabel("Gray level")
plt.ylabel("Number of pixels")
plt.xlim([0.0, 256.0])  # 设置x轴的范围
plt.plot(bin_edges[0:-1], histogram)  # 注意这里用的是bin_edges的前256个值，因为np.histogram返回的bin_edges比histogram多一个值




# 保存图表；请在这里更改为你希望保存的路径和文件名
output_path = 'misc/hist/histm_encrypt_image_peper256.png'
plt.savefig(output_path)

plt.show()
```

```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
import cv2 

def hist(img,output_path,flag=True):
    #img=cv2.imread(img)# 绘制直方图plt.figure()
    img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
    
    #plt.title("Grayscale Histogram")
    #plt.xlabel("Gray level")
    #plt.ylabel("Number of pixels")
    plt.xlim([0.0, 256.0])  # 设置x轴的范围
    if flag==False:
        plt.ylim([0.0, 650.0])  # 设置x轴的范围
    #plt.plot(bin_edges[0:-1], histogram)  # 注意这里用的是bin_edges的前256个值，因为np.histogram返回的bin_edges比histogram多一个值
    plt.hist(img.flatten(order='C'),bins=range(257),color='blue')  

    # 保存图表；请在这里更改为你希望保存的路径和文件名
    #output_path = 'misc/hist/histm_encrypt_image_baboon256.png'
    plt.savefig(output_path)
```
